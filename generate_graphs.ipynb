```python
def instance_to_graph(tsp_path, graph_path):
    with open(tsp_path, 'r') as f:
        text = f.read().replace('EOF', '')
        problem = tsplib95.parse(text)
        print('Generating graph for instance ', tsp_path)
        G = problem.get_graph()
    with open(graph_path, 'w') as f:
        print('saving to ', graph_path)
        pickle.dump(G, f, protocol=pickle.HIGHEST_PROTOCOL)
```
